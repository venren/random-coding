```python
# Plot the histogram of the samples
np.random.seed(42)
normal_samples =np.random.normal(size=10000)


plt.figure(figsize=(8, 6))
plt.hist(normal_samples, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)

# Plot the theoretical normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 100, 15)
plt.plot(x, p, 'k', linewidth=2)

# Titles and labels
plt.title('Histogram of Normal Distribution (mean=100, std=15)')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 0           # Mean of normal distribution
std = 1            # Standard deviation of normal distribution
df = 10            # Degrees of freedom for Chi-square distribution
size = 10000       # Number of samples to generate

# Generate random samples
np.random.seed(42)
normal_samples = np.random.normal(mean, std, size)       # Z: Standard normal
chi_square_samples = np.random.chisquare(df, size)       # V: Chi-square with df degrees of freedom

# Compute t-distribution
t_samples = normal_samples / np.sqrt(chi_square_samples / df)

# Plot the t-distribution
plt.figure(figsize=(8, 6))
plt.hist(t_samples, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f"t-Distribution with df={df}")
plt.xlabel("t-value")
plt.ylabel("Density")
plt.show()
```

```python
print([kurtosis(normal_samples), kurtosis(t_samples), kurtosis(chi_square_samples)])
```
