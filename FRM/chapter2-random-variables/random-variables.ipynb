```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [-2.456,-3.388,-6.816,1.531,1.737,-1.254,-1.164,1.532,2.550,0.296,-0.979,-4.259,2.810,-1.608,-0.575]
print(np.sort(data))
def printStat(data):
    m = np.mean(data)
    sd = np.std(data)
    var = sd ** 2
    normalized = normalizeData(data)
    skew = np.sum(np.array(normalized) ** 3) / len(normalized)
    kurtosis = np.sum(np.array(normalized) ** 4) / len(normalized)
    print(f"mean : {round(m,3)}")
    print(f"sd : {round(sd,3)}")
    print(f"var : {round(var,3)}")
    print(f"skew : {round(skew,3)}")
    print(f"kurtosis : {round(kurtosis,3)}")
    
def normalizeData(data):
    m = np.mean(data)
    sd = np.std(data)
    normalized = [(x-m)/sd for x in data]
    return normalized

def plot(data):
    mean = np.mean(data)
    variance = np.var(data)
    std_dev = np.std(data)
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)
    plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean:.2f}')
    plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1.5, label=f'+1 SD: {mean + std_dev:.2f}')
    plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1.5, label=f'-1 SD: {mean - std_dev:.2f}')
    plt.title('Distribution with Mean, Variance, and Standard Deviation', fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.legend()
    plt.show()   
    
printStat(data)
plot(normalizeData(data))
plot(data)

```

1) This dat has negative skew and positive kurtosis. Negative skew represents large negative values are present and tail is longer in negative side
2) Positive kurtosis means the tail is bigger.

Combining (1) and (2) - large negative values are possible
